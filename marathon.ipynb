# Exploratory Data Analysis of big dataset of ultra-marathon running

## Importing required Libraries

```python
import pandas as pd
import seaborn as sns
```

## Importing csv file into dataframe

```python
df= pd.read_csv('TWO_CENTURIES_OF_UM_RACES.csv')
```

```python
df.head(10)
```

```python
df.shape
```

```python
df.dtypes
```

### Cleaning Data

```python
# Only want 50km and 50 mi races of 2020
df[(df['Event distance/length'].isin(['50km','50mi'])) & (df['Year of event']>=2010)]
```

```python
# Races held in India
df[df['Event name'].str.split('(').str.get(1).str.split(')').str.get(0) == 'IND' ]
```

```python
# Combining both the filters to get the 50km and 50 mi races in India since the year 2010
df[(df['Event distance/length'].isin(['50km','50mi'])) & (df['Year of event']>=2010) & (df['Event name'].str.split('(').str.get(1).str.split(')').str.get(0)=='IND')]
```

```python
df2=df[(df['Event distance/length'].isin(['50km','50mi'])) & (df['Year of event']>=2010) & (df['Event name'].str.split('(').str.get(1).str.split(')').str.get(0)=='IND')]
```

```python
df2.shape
```

```python
# Remove (IND) from Event name
df2['Event name'].str.split('(').str.get(0)
```

```python
df2['Event name']= df2['Event name'].str.split('(').str.get(0)
```

```python
# Add column with athlete age
df2['Athlete Age']= 2020 - df2['Athlete year of birth']
```

```python
df2.head()
```

```python
# Removing h from athlete's performance column
df2['Athlete performance']=df2['Athlete performance'].str.split(' ').str.get(0)
```

```python
df2.head()
```

```python
# Drop Columns : Athlete club, Athlete country, Athlete year of birth and Athlete age category
df2= df2.drop(['Athlete club','Athlete country','Athlete year of birth','Athlete age category'], axis=1)
```

```python
df2.head()
```

```python
# Cleaning up NULL values
df2.isna().sum()
```

```python
df2= df2.dropna()
```

```python
df2.shape
```

```python
df2.isna().sum()
```

```python
## Check for Duplicates
df2[df2.duplicated()==True]
```

```python
# Reset Index
df2.reset_index(drop = True)
```

### Fix Datatypes

```python
df2.dtypes
```

```python
df2['Athlete Age']= df2['Athlete Age'].astype(int)
```

```python
df2['Athlete average speed']= df2['Athlete average speed'].astype(float)
```

### Renaming columns


```python
df2= df2.rename(columns=
                {'Year of event' : 'year',
                 'Event dates'  : 'race_day',
                 'Event name'  : 'race_name',
                 'Event distance/length' : 'race_length',
                 'Event number of finishers' : 'race_number_of_finishers',
                 'Athlete performance' : 'athlete_performance',
                 'Athlete gender' : 'athlete_gender',
                 'Athlete average speed' : 'athlete_average_speed',
                 'Athlete ID' : 'athlete_id',
                 'Athlete Age' : 'athlete_age'
})
```

```python
df2.head(1)
```

```python
# Reordering Columns
df3= df2[['race_day', 'race_name', 'race_length','race_number_of_finishers','athlete_id', 'athlete_gender', 'athlete_age', 
          'athlete_performance', 'athlete_average_speed','year']]
```

```python
df3.head(1)
```
